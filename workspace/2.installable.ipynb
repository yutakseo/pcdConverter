```python
def extract_ones_region(grid):
    """
    2D 리스트(grid)를 탐색하면서:
    - 값이 1인 셀은 1
    - 나머지(0, 2, None 등)는 전부 0
    으로 바꾼 새로운 2D 리스트를 반환.
    """
    ny = len(grid)
    nx = len(grid[0])

    out = [[0 for _ in range(nx)] for _ in range(ny)]

    for j in range(ny):
        for i in range(nx):
            v = grid[j][i]
            out[j][i] = 1 if v == 1 else 0

    return out
```

```python
import os
import pprint

def save_grid_as_py(grid, output_path: str, var_name: str = "GRID"):
    dirpath = os.path.dirname(output_path)
    if dirpath:
        os.makedirs(dirpath, exist_ok=True)

    with open(output_path, "w", encoding="utf-8") as f:
        f.write("# Auto-generated grid\n")
        f.write(f"{var_name} = ")
        pprint.pprint(grid, stream=f, width=120)
    print(f"[+] Saved grid to: {output_path}")
```

```python
# 예: combined_grid.py 안에 SEOCHO_GRID 라는 2D 리스트가 있다고 가정
from MAPS.seocho.seocho_combined_grid import SEOCHO_GRID as grid

ones_mask = extract_ones_region(grid)

save_grid_as_py(
    ones_mask,
    "/workspace/MAPS/seocho/installable_grid.py",
    var_name="ONES"
)
```
